```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
bar_container = ax.bar(['None', 'T/S Profiles', 'Bathy', 'SSH', 'Previous\nDisplacements'], [0.03502051, 0.0352717953, 0.0354546523, 0.0364824003, 0.0503356883])
plt.title('Sensitivy of Model to Chosen Parameters')
plt.xlabel('Parameter Excluded from training')
plt.ylabel('Average MSE')
plt.show();
```

```python
fig, ax = plt.subplots()
bar_container = ax.bar(['Persistence', 'Persistence\nand T/S', 'Persistence\nand Bathy', 'Persistence\nand SSH'], [0.035171937, 0.035698265,0.0356083447, 0.0335786777])
plt.title('Model Error With Given Parameters')
plt.xlabel('Parameters used for training')
plt.ylabel('Average MSE (3 trials)')
plt.show();
```
